## Cleaning the Data

We need to:
1. Drop rows with no Customer ID (we can't segment anonymous customers)
2. Remove cancelled orders (Invoice starts with 'C')
3. Remove rows where Quantity or Price is zero or negative
4. Parse InvoiceDate as a proper date

```python
# Drop rows missing Customer ID
df = df.dropna(subset=['Customer ID'])
print('After dropping missing Customer ID:', df.shape[0], 'rows')
```

```python
# Remove cancelled orders (Invoice number starts with 'C')
df = df[~df['Invoice'].astype(str).str.startswith('C')]
print('After removing cancellations:', df.shape[0], 'rows')
```

```python
# Remove rows with zero or negative Quantity or Price
df = df[(df['Quantity'] > 0) & (df['Price'] > 0)]
print('After removing bad Quantity/Price:', df.shape[0], 'rows')
```

```python
# Parse InvoiceDate as a datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
print('Date range:', df['InvoiceDate'].min(), 'to', df['InvoiceDate'].max())
```

```python
# Add a Revenue column (Quantity x Price per line item)
df['Revenue'] = df['Quantity'] * df['Price']
print('Total revenue in dataset: $', round(df['Revenue'].sum(), 2))
```

```python
# Final shape after all cleaning
print('Final clean dataset:', df.shape[0], 'rows,', df.shape[1], 'columns')
df.head()
```
